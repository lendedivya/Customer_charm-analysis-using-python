# converted 0 and 1 values of senior citizen to yes/no to make the esiar to understand

```python
def canv(values):
    if values == 1:
        return "yes"
    else:
        return "No"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(canv)
```

```python
# Crosstab (count)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentage
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot
ax = ct_percent.plot(kind='bar', stacked=True, figsize=(6,4))

plt.title("100% Stacked Column Chart - Churn by SeniorCitizen")
plt.ylabel("Percentage (%)")
plt.xlabel("SeniorCitizen")
plt.xticks(rotation=0)
plt.legend(title="Churn")

# ---- Add percentage labels inside bars ----
for i in range(len(ct_percent)):
    cumulative = 0
    for j in range(len(ct_percent.columns)):
        value = ct_percent.iloc[i, j]
        if value > 0:
            ax.text(i, 
                    cumulative + value/2, 
                    f'{value:.1f}%', 
                    ha='center', 
                    va='center',
                    color='white',
                    fontsize=9)
            cumulative += value

plt.show()
```

```python
# List of columns
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set figure size
plt.figure(figsize=(22, 15))

# Loop through columns
for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)   # 3 rows, 3 columns layout
    sns.countplot(x=col, data=df,hue="Churn")
    plt.title(col)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

#Customers who do not use services like Online Security, Tech Support, and Device Protection are more likely to leave the company.Customers using Fiber Internet also leave more compared to DSL users.Overall, customers who use extra services are less likely to churn.
